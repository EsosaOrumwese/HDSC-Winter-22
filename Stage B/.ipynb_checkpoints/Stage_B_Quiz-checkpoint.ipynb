12.From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two d.p.?

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
predictors = ['T2']
outcome = 'T6'

model = LinearRegression()
model.fit(energyData[predictors], energyData[outcome])
```

```python
fitted = model.predict(energyData[predictors])
r2 = r2_score(energyData[outcome], fitted)

print(f'R^2: {r2:.2f}')
```

Normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. The target variable is “Appliances”. Use a 70-30 train-test set split with a random state of 42 (for reproducibility). Run a multiple linear regression using the training set and evaluate your model on the test set. Answer the following questions:

```python
# First removing the columns ['date', 'lights']
energyData.drop(['date','lights'], axis=1, inplace=True)
```

```python
energyData.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Normalizing the data
scaler = MinMaxScaler()
energyData_norm = pd.DataFrame(scaler.fit_transform(energyData), columns=energyData.columns)
```

```python
# Using 70-30 train-test set split with a random state of 42
from sklearn.model_selection import train_test_split

predictors = ['T1', 'RH_1', 'T2', 'RH_2', 'T3', 'RH_3', 'T4', 'RH_4',
       'T5', 'RH_5', 'T6', 'RH_6', 'T7', 'RH_7', 'T8', 'RH_8', 'T9', 'RH_9',
       'T_out', 'Press_mm_hg', 'RH_out', 'Windspeed', 'Visibility',
       'Tdewpoint', 'rv1', 'rv2']
outcome = 'Appliances'

x_train, x_test, y_train, y_test = train_test_split(energyData_norm[predictors], energyData_norm[outcome],
                                                   test_size=0.3, random_state=42)
model_norm = LinearRegression()
model_norm.fit(x_train, y_train)

fitted_norm = model_norm.predict(x_test)
```

13.What is the Mean Absolute Error (in two decimal places)?

```python
from  sklearn.metrics import  mean_absolute_error, r2_score, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, fitted_norm)

print(f'Mean Absolute Error: {MAE:.2f}')
```

14.What is the Residual Sum of Squares (in two decimal places)?

```python
rss = np.sum(np.square(y_test - fitted_norm))
print(f'Residual Sum of Squares: {rss:.2f}')
```

15.What is the Root Mean Squared Error (in three decimal places)?

```python
RMSE = np.sqrt(mean_squared_error(y_test, fitted_norm))
print(f'Root Mean Squared Error: {RMSE:.3f}')
```

16.What is the Coefficient of Determination (in two decimal places)?

```python
R2_norm = r2_score(y_test, fitted_norm)

print(f'Coefficient of Determination: {MAE:.2f}')
```

17.Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
def get_wieghts_df(model, feature, col_name):
    weights = pd.Series(model.coef_, feature.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name] = weights_df[col_name]
    return weights_df

weights_lm = get_wieghts_df(model_norm, x_train, 'Linear Model Weight')
weights_lm
```

18.Train a ridge regression model with an alpha value of 0.4. Is there any change to the root mean squared error (RMSE) when evaluated on the test set?

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)

fitted_ridge = ridge_reg.predict(x_test)

RMSE_ridge = np.sqrt(mean_squared_error(y_test, fitted_ridge))
print(f'Root Mean Squared Error: {RMSE:.3f}')
```

19.Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

fitted_lasso = lasso_reg.predict(x_test)
```

```python
weights_lasso = get_wieghts_df(lasso_reg, x_train, 'Lasso Weight')
weights_lasso[weights_lasso['Lasso Weight']!=0]
```

20.What is the new RMSE with the lasso regression? (Answer should be in three (3) decimal places)

```python
RMSE_lasso = np.sqrt(mean_squared_error(y_test, fitted_lasso))
print(f'Root Mean Squared Error: {RMSE:.3f}')
```
